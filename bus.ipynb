# 버스정류장 데이터(.csv) 불러오기

```python
df_bus = pd.read_csv('C:/Users/Soobin Kim/Desktop/파이썬 프로젝트 데이터/교통시설/서울특별시 버스정류소 위치정보_UTF8.csv', names=['no', 'name', 'x_coor', 'y_coor'], header=0)
```

```python
df_bus.shape
```

```python
df_bus.head()
```

```python
import numpy as np
```

```python
coor_list = list(zip(df_bus.x_coor, df_bus.y_coor))
```

```python
addr_list = []
for coor in coor_list:
    xcoor, ycoor = coor
    url = f"https://dapi.kakao.com/v2/local/geo/coord2address.json?x={xcoor}&y={ycoor}&input_coord=WGS84"
    headers = {"Authorization": "KakaoAK 1c538cd71e5909fc30544429790dc04d"}
    api_test = requests.get(url,headers=headers)
    url_text = json.loads(api_test.text)
    if len(url_text['documents']) == 1:
        address = url_text['documents'][0]['address']['address_name']
    else: 
        address = None
    addr_list.append(address)
    print(addr_list)
```

```python
df_bus['addr'] = addr_list
print(df_bus)
```

```python
df_bus.to_csv("adde_updated.csv")
```
